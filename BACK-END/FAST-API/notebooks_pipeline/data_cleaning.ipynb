 1. 
 Fusionner 'reviewCreatedVersion' et 'appVersion'
 S'il manque la version dans 'reviewCreatedVersion', on prend 'appVersion'

 ----------------------------------------------------------------

```python
df['reviewCreatedVersion'] = df['reviewCreatedVersion'].fillna(df['appVersion'])
```

2. Supprimer 'appVersion' (devenu redondant)

```python
df.drop(columns=['appVersion'], inplace=True)
```

3. Supprimer les colonnes complÃ¨tement inutiles ou majoritairement vides
    (userImage, replyContent, repliedAt)

```python
df.drop(columns=['userImage', 'replyContent', 'repliedAt'], inplace=True)
```

4. Convertir 'score' et 'thumbsUpCount' en entier (int)
    (on remplace d'abord les Ã©ventuels NaN par 0)

```python
df['score'] = df['score'].fillna(0).astype(int)
df['thumbsUpCount'] = df['thumbsUpCount'].fillna(0).astype(int)
```

5. Convertir la colonne 'at' en format datetime, en ignorant les erreurs

```python
df['at'] = pd.to_datetime(df['at'], errors='coerce')
```

6. supprimer les lignes qui ont 'at' = NaT

```python
df = df.dropna(subset=['at'])
```

7. Supprimer les lignes qui ont encore des cellules vides
   Attention: dropna() va supprimer TOUTES les lignes qui ont au moins un NaN

```python
df.dropna(inplace=True)
```
